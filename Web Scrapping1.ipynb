```python

URL = "https://www.booking.com/searchresults.en-gb.html?ss=New+Delhi&checkin=2026-01-28&checkout=2026-02-03&group_adults=3&no_rooms=1"

MAX_RECORDS = 1500
SCROLL_PAUSE = 5

driver = webdriver.Chrome()
driver.get(URL)

time.sleep(5)

last_count = 0

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(SCROLL_PAUSE)

    soup = BeautifulSoup(driver.page_source, "lxml")
    hotel_divs = soup.find_all("div", role="listitem")

    current_count = len(hotel_divs)
    print(f"Hotels loaded: {current_count}")

    if current_count >= MAX_RECORDS:
        break

    if current_count == last_count:
        break

    last_count = current_count
```

```python
data = []

# 1. Define helper functions OUTSIDE the loop for efficiency
def safe(hotel_container, tag, cls):
    el = hotel_container.find(tag, class_=cls)
    return el.text.strip() if el else "N/A"

def safe_split(text, index, default):
    if text == "N/A" or not text:
        return default
    try:
        return text.split()[index]
    except (IndexError, AttributeError):
        return default

# 2. The Main Loop
for hotel in hotel_divs[:MAX_RECORDS]:
    
    # --- Step A: Raw Extraction ---
    # We save these to variables so we can use them in logic
    current_price = safe(hotel, "span", "b87c397a13 f2f358d1de ab607752a2").replace("₹","").strip()
    raw_original = safe(hotel, "span", "fff1944c52 d68334ea31 ab607752a2").replace("₹","").strip()

    # --- Step B: Applied Logic ---
    # If original price is N/A, we set it to the current price
    final_original_price = raw_original if raw_original != "N/A" else current_price

    # --- Step C: Create the Record ---
    record = {
        "hotel_name": safe(hotel, "div", "b87c397a13 a3e0b4ffd1"),
        "location": safe(hotel, "span", "d823fbbeed f9b3563dd4"),
        "price": current_price,
        "original_price": final_original_price,
        "ratings": safe(hotel, "div", "f63b14ab7a f546354b44 becbee2f63"),
        "score": safe_split(
            safe(hotel, "div", "f63b14ab7a dff2e52086"),
            -1,
            "0.0"
        ),
        "reviews": safe_split(
            safe(hotel, "div", "fff1944c52 fb14de7f14 eaa8455879"),
            -2,
            "0"
        ),
        "rooms": safe(hotel, "h4", "fff1944c52 f254df5361"),
        "beds": safe(hotel, "div", "ba43cae375"),
        "breakfast": safe(hotel, "span", "d651523034"),
        "transportation": safe(hotel, "span", "beb5ef4fb4"),
        "metro_access": "Metro access" if hotel.find(string=lambda x: x and "Metro access" in x) else "No Metro",
        "cancellation_policy": safe(hotel, "div", "f6e3a11b0d a19a26a18c d8df6227ee e95943ce9b f321d6ad98"),
        "center_distance": next(
            (txt for txt in hotel.stripped_strings if "km from centre" in txt.lower()),
            "N/A"
        ),
        "tax_charges": next(
            (txt.replace("+","").replace("₹","").split()[0]
             for txt in hotel.stripped_strings if "taxes and charges" in txt.lower()),
            "0"
        )
    }

    # 4. Append inside the loop
    data.append(record)

# 5. Close driver outside the loop
driver.quit()
```

```python
data = []
# 1. Initialize a set to track unique hotel names
seen_names = set() 

# Helper functions
def safe(hotel_container, tag, cls):
    el = hotel_container.find(tag, class_=cls)
    return el.text.strip() if el else "N/A"

def safe_split(text, index, default):
    if text == "N/A" or not text:
        return default
    try:
        return text.split()[index]
    except (IndexError, AttributeError):
        return default

# 2. The Main Loop
for hotel in hotel_divs[:MAX_RECORDS]:
    
    # --- Step A: Identity Check (The Duplication Filter) ---
    name_tag = hotel.find('div', class_="b87c397a13 a3e0b4ffd1")
    hotel_name = name_tag.text.strip() if name_tag else "Unknown Hotel"

    # If we have seen this hotel name before, skip to the next one
    if hotel_name in seen_names or hotel_name == "Unknown Hotel":
        continue
    
    # Otherwise, add it to our 'seen' list
    seen_names.add(hotel_name)

    # --- Step B: Raw Extraction ---
    # Standardizing numeric values by removing commas for easier math later
    current_price = safe(hotel, "span", "b87c397a13 f2f358d1de ab607752a2").replace("₹","").replace(",","").strip()
    raw_original = safe(hotel, "span", "fff1944c52 d68334ea31 ab607752a2").replace("₹","").replace(",","").strip()

    # --- Step C: Applied Logic ---
    final_original_price = raw_original if raw_original != "N/A" else current_price

    # --- Step D: Create the Record ---
    record = {
        "hotel_name": hotel_name,
        "location": safe(hotel, "span", "d823fbbeed f9b3563dd4"),
        "price": current_price,
        "original_price": final_original_price,
        "ratings": safe(hotel, "div", "f63b14ab7a f546354b44 becbee2f63"),
        "score": safe_split(safe(hotel, "div", "f63b14ab7a dff2e52086"), -1, "0.0"),
        "reviews": safe_split(safe(hotel, "div", "fff1944c52 fb14de7f14 eaa8455879"), -2, "0"),
        "rooms": safe(hotel, "h4", "fff1944c52 f254df5361"),
        "beds": safe(hotel, "div", "ba43cae375"),
        "breakfast": safe(hotel, "span", "d651523034"),
        "transportation": safe(hotel, "span", "beb5ef4fb4"),
        "metro_access": "Metro access" if hotel.find(string=lambda x: x and "Metro access" in x) else "No Metro",
        "cancellation_policy": safe(hotel, "div", "f6e3a11b0d a19a26a18c d8df6227ee e95943ce9b f321d6ad98"),
        "center_distance": next((txt for txt in hotel.stripped_strings if "km from centre" in txt.lower()), "N/A"),
        "tax_charges": next((txt.replace("+","").replace("₹","").replace(",","").split()[0] for txt in hotel.stripped_strings if "taxes and charges" in txt.lower()), "0")
    }

    data.append(record)

driver.quit()

print(f"Total Unique Hotels Collected: {len(data)}")
```
